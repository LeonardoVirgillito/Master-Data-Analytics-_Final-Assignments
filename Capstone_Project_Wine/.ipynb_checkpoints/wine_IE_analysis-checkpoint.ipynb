Grafici per produzione annuale per maggiori produttori

```python
years = oiv_wine_prod['Year'].unique()

best_prod = pd.DataFrame(columns=['Region/Country','Year','Sum of Quantity'])

i = 3
for year in years:
    bp = ((oiv_wine_prod[oiv_wine_prod['Year'] == year ]).sort_values(by=['Sum of Quantity'], ascending = False)[['Region/Country', 'Year', 'Sum of Quantity']].head(i))
    frames = [best_prod, bp]
    best_prod = pd.concat(frames)
    
```

```python
#creo un dizionario vuoto aggiungendo come chiave gli stati d'interesse
states = best_prod['Region/Country'].unique()

my_dict = {}

for s in states:
    my_dict[s] = []
print(my_dict)
```

```python
p = best_prod['Sum of Quantity'].loc[(best_prod['Year'] == 1995) & (best_prod['Region/Country'] == 'asd')]

print(len(p))
```

```python
for y in years:
    for s in states:
        p = best_prod['Sum of Quantity'].loc[(best_prod['Year'] == y) & (best_prod['Region/Country'] == s)].to_string(index=False)
        if len(p) == 0:
            p = 0
            my_dict[s].append(0)
            break
        #p = best_prod['Sum of Quantity'].loc[(best_prod['Year'] == y) & (best_prod['Region/Country'] == s)].to_string(index=False)
        my_dict[s].append(int(p))
```

```python
begin = 0
end = 4
for j in range(0, 7):
    X = years.tolist()
    X = X[begin:end]
    X_axis = np.arange(len(X))
    i = 0

    for el in my_dict:
        print(el, ': ', my_dict[el][begin:end])
        plt.bar(X_axis + i, my_dict[el][begin:end], 0.2, label = el)
        i = i + 0.2
    plt.xticks(X_axis, X)
    plt.xlabel("State for year")
    plt.ylabel("Product per 1000ha")
    plt.title("Wine Production")
    plt.legend()
    plt.show()
    
    begin = begin + 4
    end = end + 4
```
