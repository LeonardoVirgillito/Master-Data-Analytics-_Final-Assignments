### Analisi del Df Wine Reviews

```python
w_wine_mag.head()
```

```python
#Country con il Points più alto sul Vino
rating_wine_country= w_wine_mag.groupby('Country')['Points'].sum().reset_index(drop=False)
rating_wine_country.sort_values(by='Points', ascending=False).head(7)
```

```python
#Province con il Points più alto sul Vino
rating_wine_province= w_wine_mag.groupby('Province')['Points'].sum().reset_index(drop=False)
rating_wine_province.sort_values(by='Points', ascending=False).head(7)
```

```python
#Country con il costo totale dei vini
price_wine_country= w_wine_mag.groupby('Country')['Price'].sum().reset_index(drop=False)
price_wine_country.sort_values(by='Price', ascending=False).head(7)
```

```python
#Province con il Costo totale sui vini
price_wine_province= w_wine_mag.groupby('Province')['Price'].sum().reset_index(drop=False)
price_wine_province.sort_values(by='Price', ascending=False).head(7)
```

```python
# Vino con il punteggio piu alto
best_wine= w_wine_mag.groupby(['Country','Province','Winery','Variety','Price'])['Points'].sum().reset_index(drop=False)
best_wine.sort_values(by='Points', ascending=False).head(7)
```

```python
# Vino con il Price piu alto
expensive_wine= w_wine_mag.groupby(['Country','Province','Winery','Variety','Points'])['Price'].sum().reset_index(drop=False)
expensive_wine.sort_values(by='Price', ascending=False).head(7)
```

```python
'''In generale, sia il rapporto Price/Rating che Rating/Price possono essere utili per valutare la convenienza dei vini, 
ma dipende dall'obiettivo dell'analisi. Se l'obiettivo è valutare la convenienza del vino rispetto al suo prezzo,
allora il rapporto Price/Rating è il più appropriato.
Se invece si vuole valutare il valore del vino rispetto alla sua qualità, allora il rapporto Rating/Price può essere più indicato.
Se un vino ha un rating alto ma un prezzo basso, allora il rapporto Price/Rating sarà basso, indicando un'ottima convenienza del vino.
Al contrario, se un vino ha un rating basso ma un prezzo elevato, allora il rapporto Price/Rating sarà alto,
indicando una scarsa convenienza del vino.
'''
```

```python
w_wine_mag["Rating price ratio"] = w_wine_mag["Price"] / w_wine_mag["Points"]
w_wine_mag.head()
```
